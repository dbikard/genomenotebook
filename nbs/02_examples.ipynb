```python
#| export

from nbdev_genome.browser import GenomeBrowser
from bokeh.plotting import show, figure
from bokeh.layouts import column
from Bio import SeqIO
import pyBigWig
import numpy as np
```

```python
refname="NC_000913"
pos=int(0.5e6)#int(1.8e6)
win=10000
bw_file_path="data/ChIP-ACCCA-1.bw"
genome_path="data/MG1655_U00096.fasta"
gff_path="data/MG1655_U00096.gff3"
ref=next(SeqIO.parse(genome_path,"fasta"))
ref_size=len(ref.seq)
```

```python

#Importing some coverage data from a BigWig file
with pyBigWig.open(bw_file_path) as bw:
    cov=bw.values(refname,0,ref_size,numpy=True)
    cov=cov.astype('int')
    
x=np.arange(0,ref_size,10)
y=cov[::10]

g=GenomeBrowser(genome_path=genome_path, gff_path=gff_path, bounds=(0,100000))
p=g.add_track(x,y)
g.show()
    
```
